```python
######### check images of 2D features #########
def print_2D_image(mm, name):
    #length = int(mm.shape[-1]/4)
    #mm = mm[:, length : length*3, length : length*3]

    X = np.arange(-mm.shape[-1]//2, mm.shape[-1]//2)
    Y = np.arange(-mm.shape[-2]//2, mm.shape[-2]//2)
    X, Y = np.meshgrid(X, Y)
    
    print ("{} of {} save to file!".format(mm.shape[0], name))
    for i in range(mm.shape[0]):
        fig = plt.figure(figsize=(8,8))
        ax = fig.add_subplot(111, projection='3d')
        ax.set_title(name, fontsize=22)
        ax.set_xlabel(r'$f$', fontsize=22)
        ax.set_ylabel(r'$\alpha$', fontsize=22)
        ax.set_zlabel('abs', fontsize=22)
        ax.plot_surface(X, Y, mm[i,:,:]/ np.max(mm[i,:,:]), cmap=cm.viridis)
        #fig.savefig("{}_center_{}.jpg".format(name, i))
        plt.show()

print_2D_image(x2['SCD_graph'][:,3,:,:], 'SCD_graph')
print_2D_image(x2['CCSD_graph'][:,3,:,:], 'CCSD_graph')
```
