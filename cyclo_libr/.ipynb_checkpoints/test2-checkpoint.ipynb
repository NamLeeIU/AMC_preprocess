```python
import numpy as np
import lib.fot as fot
import lib.sot as crr
import lib.hos as hos
#import jingyi as jy
#%matplotlib notebook
import h5py, pickle
import numpy
import matplotlib.pyplot as plt
from scipy import signal
from scipy.fft import fft, fftfreq, fftshift
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
'''
large = 
params = { 'legend.fontsize' : large,
           'figure.figsize'  : (20, 20),
           'axes.labelsize'  : large,
           'axes.titlesize'  : large,
           'xtick.labelsize' : large,
           'ytick.labelsize' : large }
plt.rcParams.update(params)
'''
# import dataset.
mods_ = ['000','OOK', '4ASK', '8ASK', 'BPSK', 'QPSK', '8PSK', '16PSK', '32PSK', '16APSK',\
         '32APSK', '64APSK', '128APSK', '16QAM', '32QAM', '64QAM', '128QAM', \
         '256QAM', 'AM-SSB-WC', 'AM-SSB-SC', 'AM-DSB-WC', 'AM-DSB-SC', 'FM', 'GMSK', 'OQPSK']
num = [4, 5, 15, 16]

hf = h5py.File('../dataset/201801a_data_test.h5', 'r+')
x = hf['test']
pp = len(x) / 24
snr = int(pp/26) * 2
bia = 20
```

```python
mm0 = []
mm1 = []
for mod in range(1, 25):
    if mod not in num:
        continue
    print (mods_[mod])
    base = int(pp * mod - snr - bia)
    #s = x[base,:,0] + 1j*x[base,:,1]
    #print (s)
    #mm0.append(jy.SCD(s, N=256, L=16))
    s = x[base:(base+2),:,0] + 1j*x[base:(base+2),:,1]
    #s = np.expand_dims(s, axis=0)
    mm1.append(crr.SCD(s, N=512, L=16))
mm1 = np.asarray(mm1)
print (mm1.shape)
```

```python
%matplotlib inline
#512 256 156 356
#m = mm1[:,0,166:346,166:346]
#m = mm1[:,0,78:178,78:178]
m = mm1[:,0,:,:]

X = np.arange(-m.shape[-1]//2, m.shape[-1]//2)
Y = np.arange(-m.shape[-2]//2, m.shape[-2]//2)
X, Y = np.meshgrid(X, Y)

for i in range(m.shape[0]):
    print ('>>>')
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(111, projection='3d')
    ax.legend()
    ax.set_xlabel(r'$f$', fontsize=22)
    ax.set_ylabel(r'$\alpha$',fontsize=22)
    ax.set_zlabel('SCD',fontsize=22)
    ax.plot_surface(X, Y, m[i,:,:]/ np.max(m[i,:,:]), cmap=cm.viridis)
    plt.show()
```
