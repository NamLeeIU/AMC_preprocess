```python
import numpy as np
import lib.fot as fot
import lib.sot as crr
import lib.hos as hos
#%matplotlib notebook
import h5py, pickle
import numpy
import matplotlib.pyplot as plt
from scipy import signal
from scipy.fft import fft, fftfreq, fftshift

large = 18
params = { 'legend.fontsize' : large,
           'figure.figsize'  : (10, 2),
           'axes.labelsize'  : large,
           'axes.titlesize'  : large,
           'xtick.labelsize' : large,
           'ytick.labelsize' : large }
plt.rcParams.update(params)

# import dataset.
mods_ = ['000','OOK', '4ASK', '8ASK', 'BPSK', 'QPSK', '8PSK', '16PSK', '32PSK', '16APSK',\
         '32APSK', '64APSK', '128APSK', '16QAM', '32QAM', '64QAM', '128QAM', \
         '256QAM', 'AM-SSB-WC', 'AM-SSB-SC', 'AM-DSB-WC', 'AM-DSB-SC', 'FM', 'GMSK', 'OQPSK']
num = [4, 5, 15, 16]

hf = h5py.File('../dataset/201801a_data_test.h5', 'r+')
x = hf['test']
pp = len(x) / 24
snr = int(pp/26) * 2
bia = 10
```

```python
mm = []
for mod in range(1, 25):
    if mod not in num:
        continue
    out = []
    print (mods_[mod])
    for i in range(bia, bia+1):
        out.append(x[(pp*mod - snr - i),:,:])
    mm.append(out)
mm = np.asarray(mm)
print (mm.shape)
```

```python
alpha = 1.0
for n in range(mm.shape[0]):
    for c in range(mm.shape[1]):
        for i in range(mm.shape[-1]):
            plt.plot(mm[n,c,:,i], alpha=alpha)
        plt.title(mods_[num[n]] + '_time_domain')
        plt.show()
```

```python
mm = []
for mod in range(1, 25):
    if mod not in num:
        continue
    out = []
    print (mods_[mod])
    for i in range(bia, bia+1):
        out.append(fot.FT(x[(pp*mod - snr - i),:,0] + 1j*x[(pp*mod - snr - i),:,1]))
    mm.append(out)
mm = np.asarray(mm)
print (mm.shape)
mm = abs(mm)
alpha = 1.0
for n in range(mm.shape[0]):
    for c in range(mm.shape[1]):
        plt.plot(mm[n,c,:], alpha=alpha)
    plt.title(mods_[num[n]] + '_abs(FFT)')
    plt.show()
```

```python
mm = []
for mod in range(1, 25):
    if mod not in num:
        continue
    out = []
    print (mods_[mod])
    for i in range(bia, bia+1):
        s = x[(pp*mod - snr - i),:,0] + 1j*x[(pp*mod - snr - i),:,1]
        _, __, data = signal.stft(s, noverlap=128, return_onesided=False)
        data = np.abs(fftshift(data))
        data = np.mean(data, axis=-1)
        out.append(data)
    mm.append(out)
mm = np.asarray(mm)
print (mm.shape)

alpha = 1.0
for n in range(mm.shape[0]):
    for c in range(mm.shape[1]):
        plt.plot(mm[n,c,:], alpha=alpha)
    plt.title(mods_[num[n]] + '_abs(STFT)')
    plt.show()
```

### test Fourier and wavelet first order lib.

```python
waveletname = 'gaus1' # haar
mm = []
for mod in range(1, 25):
    if mod not in num:
        continue
    out = []
    print (mods_[mod])
    for i in range(bia, bia + 5):
        s = x[(pp*mod - snr - i),:,0] + 1j*x[(pp*mod - snr - i),:,1]
        s = np.asarray(s)
        a = s.real
        b = s.imag
        sn = s / np.sqrt(np.mean(a**2 + b**2))
        (data, coeff_d) = fot.CWT(s, waveletname, 256)
        (datan, coeff_d) = fot.CWT(sn, waveletname, 256)
        out.append(np.stack((data[-200,:], datan[-200, :])))
    mm.append(out)
mm = np.asarray(mm)
print (mm.shape)
hf.close()
```

```python
params = {'figure.figsize'  : (15, 2)}
plt.rcParams.update(params)
m = abs(mm)
#m = mm.real
alpha = 0.5
for n in range(m.shape[0]):
    for i in [2]:
        plt.plot(m[n,i,0,:], alpha=alpha)
        plt.plot(m[n,i,1,:], alpha=alpha)
    plt.title(mods_[num[n]])
    plt.show()
    
```
