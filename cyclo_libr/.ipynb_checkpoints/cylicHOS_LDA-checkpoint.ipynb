```python
#%matplotlib notebook
import h5py, pickle
import numpy as np
import numpy
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.linear_model import LogisticRegression as LRG
from sklearn.decomposition import PCA
import lib.hos as hos

mods_ = ['OOK', '4ASK', '8ASK', 'BPSK', 'QPSK', '8PSK', '16PSK', '32PSK', '16APSK',\
         '32APSK', '64APSK', '128APSK', '16QAM', '32QAM', '64QAM', '128QAM', \
         '256QAM', 'AM-SSB-WC', 'AM-SSB-SC', 'AM-DSB-WC', 'AM-DSB-SC', 'FM', 'GMSK', 'OQPSK']
mod = [i for i in range(0, 24) if i not in [17,18,19,20,21]]  # all digital mods.

hf = h5py.File('../dataset/201801a_data_test.h5', 'r+')
x = hf['test']
p = int(len(x) / 24)
pp = int(p / 26)
pts = 200
# x_train, x_test, y_train, y_test = train_test_split(xx, yy, test_size=0.15, random_state=42, stratify=yy)
```

### use PCA and LDA.
### SNR less than -5 dB.

```python
hoc = hos.HOC_cyclic(window_size=256, step=128)

snr = range(0,8)
xx = []
for i in mod:
    for j in snr:
        for k in range(2, pts + 2):
            base = i*p + j*pp + k
            s = hoc(x[base,:,0] + 1j*x[base,:,1])
            xx.append(abs(s).flatten())
    print (mods_[i])
xx = np.asarray(xx)
print (xx.shape)

yy = []
for i in range(len(mod)):
    yy.append([i] * int(pts * len(snr)))
yy = np.hstack(yy)
print (yy.shape)
```

```python
'''pca = PCA(n_components=1024)
xx = pca.fit_transform(xx)
print (xx.shape) '''
lda = LDA().fit(xx, yy)
lda.score(xx, yy)
```

### SNR -5 dB to +10 dB.

```python
hoc = hos.HOC_cyclic(window_size=256, step=128)

snr = range(8,16)
xx = []
for i in mod:
    for j in snr:
        for k in range(2, pts + 2):
            base = i*p + j*pp + k
            s = hoc(x[base,:,0] + 1j*x[base,:,1])
            xx.append(abs(s).flatten())
    print (mods_[i])
xx = np.asarray(xx)
print (xx.shape)

yy = []
for i in range(len(mod)):
    yy.append([i] * int(pts * len(snr)))
yy = np.hstack(yy)
print (yy.shape)
```

```python
lda = LDA().fit(xx, yy)
lda.score(xx, yy)
```

```python
lda = LDA(n_components=2)
lda = lda.fit(xx, yy)
X_lda = lda.transform(xx)
```

```python
colors = cm.rainbow(np.linspace(0, 1, len(mod)))
plt.figure(figsize = (7, 7))
plt.scatter(X_lda[:, 0], X_lda[:, 1], alpha=0.6, c=colors[yy])
```
